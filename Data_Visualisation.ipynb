5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 100
x = np.random.normal(0, 1, n_samples)  # Feature X
y = 2 * x + np.random.normal(0, 0.5, n_samples)  # Feature Y, correlated with X

# Create a DataFrame
data = pd.DataFrame({'Feature X': x, 'Feature Y': y})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Set the random seed for reproducibility
np.random.seed(20)

# Generate the sales data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),  # Randomly assigns months
    'Day': np.random.choice(range(1, 32), 100),  # Randomly assigns days between 1 and 31
    'Sales': np.random.randint(1000, 5000, 100)  # Randomly assigns sales figures between 1000 and 5000
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Create a pivot table to reshape the data for the heatmap
sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create a heatmap to visualize sales variation across different months and days
fig = px.imshow(sales_pivot, 
                color_continuous_scale='Viridis', 
                labels=dict(x='Month', y='Day', color='Sales'),
                title='Sales Variation Across Months and Days')

# Show the plot
fig.show()
```

```python
import numpy as np
import plotly.express as px
import pandas as pd

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)  # Create a 2D grid from x and y values

# Calculate z using the function z = sin(sqrt(x^2 + y^2))
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays for plotly express
x_flat = x.flatten()
y_flat = y.flatten()
z_flat = z.flatten()

# Create a DataFrame for plotly express
df = pd.DataFrame({'X': x_flat, 'Y': y_flat, 'Z': z_flat})

# Create a 3D surface plot using plotly express
fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Z',
                    color_continuous_scale='Viridis',
                    title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Update the layout to increase size
fig.update_layout(
    width=1000,  # Set width in pixels
    height=800   # Set height in pixels
)

# Show the plot
fig.show()
```

# BOKEH ASSIGNMENT

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                 # Sine of x

# Create a Bokeh plot
output_notebook()
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label='sin(x)', line_width=2,color="red")

# Show the plot
show(p)
```

2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Set up the output to display in the notebook
output_notebook()

# Generate random data
n = 100  # number of points
np.random.seed(0)  # for reproducibility
x = np.random.rand(n) * 10  # random x values
y = np.random.rand(n) * 10  # random y values
sizes = np.random.randint(10, 200, size=n)  # random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)  # random colors

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a scatter plot
p = figure(title="Bokeh Scatter Plot", x_axis_label='X', y_axis_label='Y', tools="pan,wheel_zoom,reset")
p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

# Show the plot
show(p)
```

```python
# Import necessary libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd

# Set up the output to display in the notebook
output_notebook()

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame
data = pd.DataFrame({'fruits': fruits, 'counts': counts})

# Create a bar chart
p = figure(x_range=data['fruits'], title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts')

# Add bars to the bar chart
p.vbar(x='fruits', top='counts', source=data, width=0.9,color="brown")

# Show the plot
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
# Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Set up the output to display in the notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)  # Generate 1000 random numbers from a normal distribution

# Create a histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure for the histogram
p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency')

# Use vbar to create the histogram
x_values = np.arange(len(hist))  # Create x values for each bin
p.vbar(x=x_values, top=hist, width=0.9, fill_color="navy", alpha=0.5)

# Show the plot
show(p)
```
